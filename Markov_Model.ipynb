# 1- Load Data 

```
locations=[]
def Load_Data(file):
  with open(file) as f:
     for line in f:
         locations.append(line.split()[1]) 
  return locations
```

# 2- Split Data ( Train/ Test)


```
def Split_sequence(locations,train_percentage):
  train_locations_index = int(train_percentage * len(locations))
  train_locations = locations[:train_locations_index]
  test_locations = locations[train_locations_index:]
  return train_locations,test_locations
```

# 3-  Markov Model

```
def train_markov_model(locations, train_percentage, w):
  d = {}
  
  for i in range(len(locations)-w):
      context = ','.join(locations[i:i+w])
      next_location = locations[i+w]
      if(context in d.keys()):
          if(next_location in d[context].keys()):
              d[context][next_location]=d[context][next_location]+1
          else:
              d[context][next_location]=1
      else:
          d[context]={next_location:1}
  
  return d
```

# 4- Test





## 4.1- Breaking Ties

```
def most_frequent(List): 
    return max(set(List), key = List.count) 
```

## 4.2-  Next Location


```
def nxt_loc(data_test,model,test_loc,w,i,method):
  if (Built_context(data_test,i,w) in model.keys() ):
     next_locations_dic = model[test_loc]
     inverse = [(value, key) for value, key in next_locations_dic.items()]
     Max_val=(max(inverse)[1])
     list_of_max=[k for k,v in next_locations_dic.items() if v == Max_val]
     if (len(list_of_max)>1):
      if (method == "first"):
        next_loc = locations[locations.index(list_of_max[0])]
      elif (method =="last"):
        next_loc = locations[locations.index(list_of_max[len(list_of_max)-1])]
      elif (method == "most"):
        next_loc=most_frequent(locations)
      return next_loc
     else:
      next_loc=max(next_locations_dic,key=next_locations_dic.get)
      return next_loc
  else:
    undefined_seq = Built_context(data_test,i,w)
    return undefined_seq
```

## 4.3- Build Context

```
def Built_context(test_locations,index,w):
  test_loc=''
  for i in range(len(test_locations[index:(index+w-1)])):
    test_loc+=test_locations[index+i]+','
  test_loc += test_locations[index+w-1]
  #print(test_loc)
  return test_loc
```

## 4.4 Comparing predicted data with real data


```
def test (test,w, dictionary):
  correct=0
  incorrect=0
  undefined=0
  for i in range(len(test)-w):
    prediction = nxt_loc(test,dictionary,Built_context(test,i,w),w,i,method="first")
    real_location = test[i+w]
    print("predict= ", prediction, "|| Real Location = ",real_location, "||Context= ",Built_context(test,i,w)) 
    if (Built_context(test,i,w) in dictionary.keys() ):
      if (prediction == real_location ):
        correct +=1
      else:
        incorrect +=1
    else:
      undefined +=1
  accuracy = correct/(len(test))
  return correct,incorrect,accuracy,undefined
```
